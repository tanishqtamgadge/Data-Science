```python
# INDEED Major project
import pandas as pd
import requests
from bs4 import BeautifulSoup
import selenium                 # automate web browser
from selenium import webdriver  # it will use browser from the machine
import time                     # halt pyhton program                 

# url generator
def get_url(position,location):
    position = position.replace(' ','+')
    location = location.replace(' ','+')
    template = 'https://in.indeed.com/jobs?q={}&l={}'
    url = template.format(position,location)
    return url

# scrapper
def get_job(card):
    try:
        job_Name = card.find('a', {'class': 'jcs-JobTitle'})
        job_Name = job_Name.get_text(strip=True) if job_Name else "Not Available"

        company_Name = card.find('span', {"data-testid": "company-name"})
        company_Name = company_Name.get_text(strip=True) if company_Name else "Not Available"

        company_location = card.find('div', {"data-testid": "text-location"})
        company_location = company_location.get_text(strip=True) if company_location else "Not Available"

        company_requirements = card.find('div', {"data-testid": "jobsnippet_footer"})
        company_requirements = company_requirements.get_text(strip=True) if company_requirements else "Not Available"

        salary = card.find('div', {"data-testid": "attribute_snippet_testid"})
        salary = salary.get_text(strip=True) if salary else "Not Available"

        links = card.find('a', {'class': 'jcs-JobTitle'})
        links = "https://in.indeed.com" + links['href'] if links else "Not Available"

    except Exception as e:
        print(f"Error extracting job details: {e}")
        return ("Not Available",) * 7  

    return job_Name, company_Name, company_location, company_requirements, salary, links

    return info

def main(position,location) :
    records = []
    url = get_url(position,location)
    
    time.sleep(3) 
    options = webdriver.ChromeOptions()
    options.add_argument("--disable-blink-features=AutomationControlled")
    driver = webdriver.Chrome(options=options)
    driver.maximize_window()
    driver.get(url)

    time.sleep(10) 
    soup = BeautifulSoup(driver.page_source,'html.parser')
    cards = soup.find_all('li','css-1ac2h1w eu4oa1w0')
    
    for i in cards :
        jobDetail = get_job(i)
        records.append(jobDetail)

    col = ["job_Name","company_Name","company_location","company_requirements","salary","links"]

    data = pd.DataFrame(records,columns=col)
    data.to_csv('D:\\the boys\\webscrapperindeed_data.csv',encoding='utf-8')
```
