```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#genearate a url for job position and location
def get_url(position,location):
    position = position.replace(' ','-')
    location = location.replace(' ','-')
    template = 'https://my.jobstreet.com/{}-jobs-jobs/in-{}'
    url = template.format(position,location)
    return url
```

```python
# fetch job information from every cards

def get_jobs(card):
    try:
        job_name = card.find('a',{'data-testid':'job-card-title'}).text
        job_url = card.find('a',{'data-testid':'job-card-title'})
        job_url = 'https://my.jobstreet.com'+job_url['href']
        cname = card.find('a',{'data-automation':'jobCompany'}).text
        job_loc = card.find('a',{'data-automation':'jobLocation'}).text
        salary = card.find('span','gepq850 _1q03wcw2 eihuid4z eihuid0 eihuidr _1q03wcw4').text
    except:
        job_name = 'Not Mentioned'
        job_url = 'Not Available'
        cname = 'Not Mentioned'
        job_loc = 'Not Mentioned'
        salary = 'Not Available'
        
    job_info = (job_name,job_url,cname,job_loc,salary)

    return job_info
```

```python
records = []
for i in cards:
    jobDetails = get_jobs(i)
    records.append(jobDetails)
```
